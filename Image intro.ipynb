# OpenCV

```python
import cv2
```

```python
img = cv2.imread('sample dataset/hummingBird.jpg')
```

## It is a good practice to check the type of image because if we specify the wrong file path we do not get error message but on checking the type of variable it would be 'None'

```python
type(img)
```

```python
img_wrong = cv2.imread('wrong/file/path')
type(img_wrong)
```

```python
img.shape
```

# Matplotlib color formatting -> RGB
# OpenCV color formatting -> BGR

```python
plt.imshow(img)
```

## converting the color format

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

## Reading an image as a grayscale

```python
img_gray = cv2.imread('sample dataset/hummingBird.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
# Doesn't show black and white image because again cv2 reads it as a bgr format
```

```python
plt.imshow(img_gray,cmap="gray")
```

# Resizing images

```python
fix_img.shape
```

```python
resize_img = cv2.resize(fix_img,(1000,500))
plt.imshow(resize_img)
```

## Resizing image with ration

```python
w_ratio = 0.3
h_ratio = 0.7
img_ratio_resize = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
plt.imshow(img_ratio_resize)
```

# Image flipping

```python
plt.imshow(cv2.flip(fix_img,0))
```

```python
plt.imshow(cv2.flip(fix_img,1))
```

```python
plt.imshow(cv2.flip(fix_img,-1))
```

## Image rotate

```python
img_r = cv2.rotate(fix_img,cv2.ROTATE_180)
plt.imshow(img_r)
```

# Image save

```python
cv2.imwrite("sample dataset/hummingBirdRotate.jpg",img_r)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.imshow(fix_img)
```
